```python
# Importo le librerie che mi servono:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Aggiungo il file csv da analizzare e controllo le prime righe
insurance_data = pd.read_csv("./insurance.csv", dtype={'region': 'category'})


# Creo una funzione per generare un boxplot
# Funzione per creare un boxplot
def boxplot_by_category(numeric_column, category_column, **kwargs):
    print(f"\nSto creando un boxplot di {numeric_column} in base a {category_column}")
    
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=category_column, y=numeric_column, data=insurance_data, palette="pastel", hue=category_column, dodge=True, legend=False)
    plt.title(f'Boxplot di {numeric_column} per {category_column}', fontsize=14)
    title = kwargs.get('title', None)
    if title:
        plt.title(title, fontsize=16)  
    plt.show()


# Esempi di utilizzo della funzione 
boxplot_by_category('charges', 'sex', title='Boxplot delle spese per sesso')
boxplot_by_category('charges', 'children', title='Boxplot delle spese per bambini')
boxplot_by_category('charges', 'smoker',title='Boxplot delle spese per fumatori')

```

```python
# Importo le librerie che mi servono:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Aggiungo il file csv da analizzare e controllo le prime righe
insurance_data = pd.read_csv("./insurance.csv", dtype={'region': 'category'})

# Tasformo i dati stinghe in valori numerici
insurance_data['sex'] = insurance_data['sex'].map({'female': 0, 'male': 1})
insurance_data['smoker'] = insurance_data['smoker'].map({'yes': 1, 'no': 0})

# Calcolo la correlazione tra BMI e costi assicurativi
correlation_bmi_charges = insurance_data['bmi'].corr(insurance_data['charges'])

# Stampo il risultato
print(f"Correlazione tra BMI e Costi Assicurativi: {correlation_bmi_charges}")

# Valuto la correlazione e stampo il risultato desiderato
if correlation_bmi_charges == 0:
    print("La correlazione  vicina a zero, indicando una scarsa relazione lineare.")
elif 0 < correlation_bmi_charges < 0.1:
    print("La correlazione  debole, indicando una relazione molto lieve.")
elif 0.1 <= correlation_bmi_charges < 0.3:
    print("La correlazione  leggera, indicando una relazione crescente moderata.")
elif correlation_bmi_charges >= 0.3:
    print("La correlazione  significativa, indicando una relazione crescente forte.")
else:
    print("La correlazione  negativa, indicando una relazione decrescente tra BMI e Costi Assicurativi.")

# Calcolo la correlazione tra Fumo e costi assicurativi
correlation_smoker_charges = insurance_data['smoker'].corr(insurance_data['charges'])

# Stampo il risultato
print(f"Correlazione tra Fumo e Costi Assicurativi: {correlation_smoker_charges}")

# Valuto la correlazione e stampo il risultato desiderato
if correlation_smoker_charges == 0:
    print("La correlazione  vicina a zero, indicando una scarsa relazione lineare.")
elif 0 < correlation_smoker_charges < 0.1:
    print("La correlazione  debole, indicando una relazione molto lieve.")
elif 0.1 <= correlation_smoker_charges < 0.3:
    print("La correlazione  leggera, indicando una relazione crescente moderata.")
elif correlation_smoker_charges >= 0.3:
    print("La correlazione  significativa, indicando una relazione crescente forte.")
else:
    print("La correlazione  negativa, indicando una relazione decrescente tra Fumo e Costi Assicurativi.")
```

```python
# Importo le librerie che mi servono:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Aggiungo il file csv da analizzare e controllo le prime righe
insurance_data = pd.read_csv("./insurance.csv", dtype={'region': 'category'})

# Tasformo i dati stinghe in valori numerici
insurance_data['sex'] = insurance_data['sex'].map({'female': 0, 'male': 1})

# Esploro la distribuzione di fumatori e non fumatori
sns.countplot(x='smoker', data=insurance_data, palette="pastel")
plt.title('Distribuzione di Fumatori e Non Fumatori')
plt.xlabel('Fumatore')
plt.ylabel('Conteggio')
plt.show()

# Calcolo la media dei costi assicurativi per fumatori e non fumatori
mean_charges_by_smoker = insurance_data.groupby('smoker')['charges'].mean()

# Stampo il risultato
print("Media dei costi assicurativi per fumatori e non fumatori:")
print(mean_charges_by_smoker)
```

```python
# Importo le librerie che mi servono:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Aggiungo il file csv da analizzare e controllo le prime righe
insurance_data = pd.read_csv("./insurance.csv", dtype={'region': 'category'})

# Tasformo i dati stinghe in valori numerici
insurance_data['sex'] = insurance_data['sex'].map({'female': 0, 'male': 1})
insurance_data['smoker'] = insurance_data['smoker'].map({'yes': 1, 'no': 0})

# Utilizzo il one-hot encoding per la colonna 'region' senza eliminare la prima colonna
insurance_data = pd.get_dummies(insurance_data, columns=['region'], drop_first=False)

# Funzione per visualizzare istogramma e statistiche di una colonna
def analyze_column(column_name):
    plt.figure(figsize=(10, 6))
    sns.histplot(insurance_data[column_name], kde=True, color="darkgreen")
    plt.title(f'Distribution for {column_name}', fontsize=14)
    plt.show()
    print(f"Statistics for {column_name}:\n{insurance_data[column_name].describe()}\n")

analyze_column('age')
analyze_column('bmi')
analyze_column('children')
analyze_column('charges')
```
